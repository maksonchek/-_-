```python
import numpy as np
def get_time(method, lists):
    import time
    times = []
    res = None
    for i in range(5):
        start = time.time()
        res = method(lists)
        end = time.time()
        times.append(end - start)
    return np.mean(times), res
```

**Брутфорс**

```python
def intersection_bruteforce(list1, list2):
    result = []

    for item1 in list1:
        for item2 in list2:
            if item1 == item2:
                result.append(item1)

    return result

def intersection_bruteforce_multiple(lists):
    result = lists[0]

    for lst in lists[1:]:
        result = intersection_bruteforce(result, lst)

    return result
```

```python
time, res = get_time(intersection_bruteforce_multiple, lists)
correct_ans = res
print(res)
print(f"Время выполнения очевидного способа: {time}")
```

**Бинарный поиск**

```python
def bin_search(mas, x, l, r):
    if mas[l] == x:
        return x
    if mas[r] == x:
        return x
    while l < r:
        m = (l+r)//2
        if mas[m] > x:
            r = m
        elif mas[m] < x:
            l = m + 1
        elif mas[m] == x:
            return x
    return None

def multipple_bin_search(lists):
    result = lists[0]

    for lst in lists[1:]:
        ans = []
        for num in lst:
            x = bin_search(result, num, 0, len(ans)-1)
            if x:
                ans.append(x)
        result = ans
    return result
```

```python
time, res = get_time(multipple_bin_search, lists)
correct_ans = res
print(res == correct_ans)
print(f"Время выполнения способа с бинарным поиском: {time}")
```

```python
def intersection_bruteforce(list1, list2):
    result = []

    for item1 in list1:
        for item2 in list2:
            if item1 == item2:
                result.append(item1)

    return result

def intersection_bruteforce_multiple(*lists):
    if not lists:
        return []

    result = lists[0]

    for lst in lists[1:]:
        result = intersection_bruteforce(result, lst)

    return result
```
